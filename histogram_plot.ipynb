# Histogram Plot

## Common uses:

Analyzing scores, ages, incomes

Visualizing frequency distribution

```python
import matplotlib.pyplot as plt
```

```python
data = [10, 20, 20, 30, 40, 40, 40, 50, 60, 70]
print(data)
```

## Histogram with Custom Bins

```python
plt.hist(data, bins=5)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Basic Histogram")
plt.show()
# Bin 1: 10–22
# Bin 2: 22–34
# Bin 3: 34–46
# Bin 4: 46–58
# Bin 5: 58–70
# X-axis → represents value ranges (called bins)
# Y-axis → represents frequency (how many data points fall into each bin)
# Bars → height = frequency
```

## Histogram with density

```python
plt.hist(data, bins=5, density=True, color='green', alpha=0.6)
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Histogram with Density")
plt.show()

# Explainations:
# How Density is Calculated
# Total number of points = 10
# Bin width = 12 (for 5 bins from 10–70 → bin edges: 10–22, 22–34, etc.)
# Count in Bin 1 = 3
# Density for Bin 1 = Count / (Total points × Bin width) = 3 / (10 × 12) ≈ 0.025
# Heights of all bins sum up to 1 when multiplied by bin width
```

## Stacked histogram 

```python
data1 = [10, 20, 20, 30, 40, 40, 40, 50, 60, 70]
data2 = [15, 25, 25, 35, 45, 45, 50, 55, 65, 75]

plt.hist([data1, data2], bins=5, color=['blue', 'orange'], stacked=True, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Stacked Histogram")
plt.show()

# Explainations:
# Example Calculation
# Bin 1 (10–22):
# data1 → 3 points
# data2 → 1 point
# Stacked height = 3 + 1 = 4
# Bin 2 (22–34):
# data1 → 1 point
# data2 → 2 points
# Height = 3
# And so on…
# Stacked bars let you see both total and individual contributions.
```

## cumulative histogram

```python
plt.hist(data, bins=5, cumulative=True, color='purple', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Cumulative Frequency")
plt.title("Cumulative Histogram")
plt.show()

# Explaination: 
# Example Calculation for Our Data
# Bins (10–22, 22–34, 34–46, 46–58, 58–70)
# Heights for cumulative histogram:
# Bin 1 → 10, 20, 20 → 3
# Bin 2 → 30 → 3 + 1 = 4
# Bin 3 → 40, 40, 40 → 4 + 3 = 7
# Bin 4 → 50 → 7 + 1 = 8
# Bin 5 → 60, 70 → 8 + 2 = 10
# Final bar heights show how many points are ≤ upper edge of each bin.

```
