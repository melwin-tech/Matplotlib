# Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [120, 150, 100, 170, 160, 180]
print(months)
print(sales)
```

### Vertical Bar Plot


```python

plt.bar(months, sales, color='skyblue', width=0.6)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales")
plt.show()
```

### Horizontal Bar Plot

```python
plt.barh(months, sales, color='orange', height=0.5)
plt.xlabel("Sales")
plt.ylabel("Months")
plt.title("Monthly Sales")
plt.show()
```

### Adding Value Labels on Bars

```python

bars = plt.bar(months, sales, color='lightgreen')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, yval, ha='center')

plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

### Stacked Bar Plot Example

```python
online_sales = [60, 80, 40, 100, 90, 110]
offline_sales = [60, 70, 60, 70, 70, 70]

plt.bar(months, online_sales, label='Online', color='blue')
plt.bar(months, offline_sales, bottom=online_sales, label='Offline', color='green')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Stacked Monthly Sales")
plt.legend()
plt.show()
```

### Grouped (Side-by-Side) Bar Plot Example

```python
import numpy as np

x = np.arange(len(months))
width = 0.35

plt.bar(x - width/2, online_sales, width, label='Online', color='blue')
plt.bar(x + width/2, offline_sales, width, label='Offline', color='green')
plt.xticks(x, months)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Grouped Monthly Sales")
plt.legend()
plt.show()
```
