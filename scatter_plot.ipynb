# Bar Plot

## Use Case

Relationship between variables

Clustering

Outlier detection

```python
import matplotlib.pyplot as plt
```

```python
x = [10, 20, 30, 40, 50, 60]
y = [12, 25, 28, 45, 48, 65]
print(x)
print(y)
```

```python
# All common configurations

plt.scatter(
    x, y,
    s=100,               # size
    color='red',             # color
    marker='o',          # marker style
    alpha=0.6,           # transparency
    edgecolors='black'
)
plt.show()
```

```python
# Variant : Bubble Chart

x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 5, 4, 6, 8]
sizes = [100, 200, 300, 400, 500, 600]  # SAME length

plt.scatter(x, y, s=sizes, alpha=0.5)
plt.show()
```
