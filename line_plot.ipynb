# Line Plot

## Use Case

Trends over time

Continuous data

Comparisons between datasets

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,4,6,5,10]
print(x)
print(y)
```

```python
# All common configurations
plt.plot(
    x, y,
    color='blue',        # line color
    linestyle='--',      # '-', '--', ':', '-.'
    linewidth=2,         # thickness
    marker='o',          # 'o', 's', '^', '*'
    markersize=8,
    markerfacecolor='red',
    label='Sales'
)
plt.show()
```

```python
# multiple line plots 

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [5, 15, 25, 35]

plt.plot(x, y1, color='blue', marker='o', label='Line 1')
plt.plot(x, y2, color='red', marker='s', label='Line 2')

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Two Line Plot")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Variant : Step plot in line plot

x = [1,2,3,4,5]
y = [2,4,6,5,10]
plt.step(x, y)
plt.show()
```

```python
# Filled Line Plot (Area Under Curve)

plt.fill_between(x, y, alpha=1, color='red')
plt.show()
```
